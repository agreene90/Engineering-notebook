# Kinematics

Forward Kinematics:
-------------------
x = L1 * cos(θ1) + L2 * cos(θ1 + θ2)
y = L1 * sin(θ1) + L2 * sin(θ1 + θ2)

Inverse Kinematics:
-------------------
θ2 = arccos((x^2 + y^2 - L1^2 - L2^2) / (2 * L1 * L2))
θ1 = arctan(y/x) - arctan((L2 * sin(θ2)) / (L1 + L2 * cos(θ2)))

Jacobian Matrix:
----------------
J = [ ∂x/∂θ1  ∂x/∂θ2 ]
    [ ∂y/∂θ1  ∂y/∂θ2 ]

where
∂x/∂θ1 = -L1 * sin(θ1) - L2 * sin(θ1 + θ2)
∂x/∂θ2 = -L2 * sin(θ1 + θ2)
∂y/∂θ1 = L1 * cos(θ1) + L2 * cos(θ1 + θ2)
∂y/∂θ2 = L2 * cos(θ1 + θ2)


# Numerical Methods

Newton-Raphson Method:
----------------------
x_(n+1) = x_n - f(x_n) / f'(x_n)

Runge-Kutta Method (4th Order):
-------------------------------
k1 = h * f(t_n, y_n)
k2 = h * f(t_n + h/2, y_n + k1/2)
k3 = h * f(t_n + h/2, y_n + k2/2)
k4 = h * f(t_n + h, y_n + k3)

y_(n+1) = y_n + (1/6)*(k1 + 2*k2 + 2*k3 + k4)

where
h = step size
t_n, y_n = current time and value
y_(n+1) = next value

## Simple Harmonic Motion

Simple Harmonic Motion (SHM) describes the oscillatory motion of a mass-spring system. The displacement \( x(t) \) is given by the equation:

\[ x(t) = A \cos(\omega t + \phi) \]

where:
- \( A \) is the amplitude
- \( \omega \) is the angular frequency
- \( \phi \) is the phase angle
- \( t \) is the time

Use the sliders below to adjust the parameters and observe the effect on the motion.


```python

import numpy as np
import matplotlib.pyplot as plt

def simple_harmonic_motion(A, omega, phi, t):
    return A * np.cos(omega * t + phi)

def plot_shm(A, omega, phi):
    t = np.linspace(0, 10, 1000)
    x = simple_harmonic_motion(A, omega, phi, t)
    
    plt.figure(figsize=(10, 5))
    plt.plot(t, x, label='x(t)')
    plt.title('Simple Harmonic Motion')
    plt.xlabel('Time (s)')
    plt.ylabel('Displacement (m)')
    plt.legend()
    plt.grid(True)
    plt.show()

from ipywidgets import interact, FloatSlider
interact(plot_shm, 
         A=FloatSlider(value=1.0, min=0.0, max=10.0, step=0.1, description='Amplitude (A)'),
         omega=FloatSlider(value=2.0, min=0.0, max=10.0, step=0.1, description='Angular Frequency (ω)'),
         phi=FloatSlider(value=0.0, min=0.0, max=2*np.pi, step=0.1, description='Phase Angle (φ)'));
```


## Kinematics

### Forward Kinematics

Forward kinematics involves calculating the position and orientation of a robot's end-effector given the joint parameters. 

### Inverse Kinematics

Inverse kinematics involves calculating the joint parameters needed to place the end-effector at a desired position and orientation.

Below are the equations and examples demonstrating both concepts.


```python

import numpy as np
import matplotlib.pyplot as plt

# Example: 2D Robot Arm with two links
def forward_kinematics(theta1, theta2, l1, l2):
    x = l1 * np.cos(theta1) + l2 * np.cos(theta1 + theta2)
    y = l1 * np.sin(theta1) + l2 * np.sin(theta1 + theta2)
    return x, y

# Parameters
theta1 = np.radians(45)
theta2 = np.radians(30)
l1 = 1.0
l2 = 0.5

x, y = forward_kinematics(theta1, theta2, l1, l2)

plt.figure(figsize=(6, 6))
plt.plot([0, l1 * np.cos(theta1)], [0, l1 * np.sin(theta1)], label='Link 1')
plt.plot([l1 * np.cos(theta1), x], [l1 * np.sin(theta1), y], label='Link 2')
plt.plot(x, y, 'ro', label='End Effector')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.title('2D Robot Arm Forward Kinematics')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)
plt.show()
```


## Control Theory

### PID Controller

A PID controller is used to maintain a desired output by adjusting the control inputs. The PID controller's output is given by:

\[ u(t) = K_p e(t) + K_i \int e(t) dt + K_d rac{de(t)}{dt} \]

where:
- \( K_p \) is the proportional gain
- \( K_i \) is the integral gain
- \( K_d \) is the derivative gain
- \( e(t) \) is the error at time \( t \)

The following interactive simulation allows you to adjust the PID parameters and observe the system's response.


```python

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, VBox, Label

class PID:
    def __init__(self, P=1.0, I=0.0, D=0.0, setpoint=0):
        self.Kp = P
        self.Ki = I
        self.Kd = D
        self.setpoint = setpoint
        self.last_error = 0.0
        self.integral = 0.0

    def compute(self, current_value):
        error = self.setpoint - current_value
        self.integral += error
        derivative = error - self.last_error
        self.last_error = error
        return self.Kp * error + self.Ki * self.integral + self.Kd * derivative

def system_response(Kp, Ki, Kd):
    pid = PID(P=Kp, I=Ki, D=Kd, setpoint=1)
    time = np.linspace(0, 10, 100)
    output = []
    error_list = []
    control_list = []
    current_value = 0
    
    for t in time:
        control = pid.compute(current_value)
        current_value += control * 0.1  # simple system dynamics
        output.append(current_value)
        error_list.append(pid.last_error)
        control_list.append(control)
    
    fig, ax = plt.subplots(3, 1, figsize=(10, 15), sharex=True)
    
    ax[0].plot(time, output, label='System Output', color='blue')
    ax[0].axhline(y=1, color='r', linestyle='--', label='Setpoint')
    ax[0].set_title('PID Controller Response')
    ax[0].set_ylabel('Output')
    ax[0].legend()
    ax[0].grid(True)
    
    ax[1].plot(time, error_list, label='Error', color='orange')
    ax[1].set_title('Error over Time')
    ax[1].set_ylabel('Error')
    ax[1].legend()
    ax[1].grid(True)
    
    ax[2].plot(time, control_list, label='Control Signal', color='green')
    ax[2].set_title('Control Signal over Time')
    ax[2].set_xlabel('Time')
    ax[2].set_ylabel('Control Signal')
    ax[2].legend()
    ax[2].grid(True)
    
    plt.tight_layout()
    plt.show()

def PID_controls():
    interact(system_response, 
             Kp=FloatSlider(value=1.0, min=0.0, max=10.0, step=0.1, description='Proportional Gain (Kp)'),
             Ki=FloatSlider(value=0.0, min=0.0, max=1.0, step=0.01, description='Integral Gain (Ki)'),
             Kd=FloatSlider(value=0.0, min=0.0, max=1.0, step=0.01, description='Derivative Gain (Kd)'));
    
    control_info = VBox([
        Label("Use the sliders above to adjust the PID controller gains and observe the effects on the system's response."),
        Label("Proportional Gain (Kp): Determines the reaction to the current error."),
        Label("Integral Gain (Ki): Determines the reaction based on the sum of recent errors."),
        Label("Derivative Gain (Kd): Determines the reaction based on the rate at which the error has been changing."),
        Label("System Output: The response of the system to the control signals."),
        Label("Error: The difference between the setpoint and the current output."),
        Label("Control Signal: The output from the PID controller that drives the system.")
    ])
    display(control_info)

PID_controls()
```


## Signal Processing

### Fourier Transform

The Fourier Transform is a mathematical transformation used to analyze the frequencies present in a signal. The Discrete Fourier Transform (DFT) is given by:

\[ X_k = \sum_{n=0}^{N-1} x_n e^{-i 2 \pi k n / N} \]

where \( x_n \) is the input signal and \( X_k \) is the DFT of the signal.

Below is an example demonstrating the Fourier Transform of a signal.


```python

import numpy as np
import matplotlib.pyplot as plt

# Generate a sample signal
Fs = 500  # Sampling rate
T = 1.0  # seconds
t = np.linspace(0, T, int(Fs*T), endpoint=False)  # time variable
# Create a signal with two frequencies
f1, f2 = 5, 50  # frequencies in Hz
signal = np.sin(2 * np.pi * f1 * t) + 0.5 * np.sin(2 * np.pi * f2 * t)

# Compute the Fourier Transform
freq = np.fft.fftfreq(len(t), 1/Fs)
signal_ft = np.fft.fft(signal)

# Plot the signal
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()

# Plot the Fourier Transform
plt.subplot(2, 1, 2)
plt.stem(freq, np.abs(signal_ft), 'b', markerfmt=" ", basefmt="-b")
plt.title('Fourier Transform')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
